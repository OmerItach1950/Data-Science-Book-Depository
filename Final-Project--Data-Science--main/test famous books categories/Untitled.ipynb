```python
def crawl_books_urls(n: int) -> list[str]:
    books_urls = []
    page_base_url = 'https://www.bookdepository.com/search?searchLang=123&page='
    book_base_url = 'https://www.bookdepository.com'

    for i in range(n):
        url = f'{page_base_url}{i + 1}'
        print('page number', i + 1, url)
        currentPage = get_soup_object(url)
        items = currentPage('div', {'class': 'book-item'})
        for book in items:
            book_url = book.find('h3').find('a')['href']
            books_urls.append(f'{book_base_url}{book_url}')  
            
    return books_urls
```

```python
def get_categories(url: str) -> set[str]:
    soup = get_soup_object(url)
    item_ol = soup.find('ol', {'class': 'breadcrumb'})
    categories: set = {}
    try:
        categories = {t.text.strip() for t in item_ol('a')}
    except:
        return {}
    return categories
```

```python
l = crawl_books_urls(200)
```

```python
popular_categories = dict()
```

```python
for i, url in enumerate(l):
    print('book number', i, 'amount of categories', len(popular_categories))
    current_categories = get_categories(url)
    for category in current_categories:
        if category in popular_categories:
            popular_categories[category] += 1
        else:
            popular_categories[category] = 0
```
